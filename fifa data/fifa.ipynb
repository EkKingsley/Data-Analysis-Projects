#### exploratory data analysis

```python
fifa.club.value_counts()[:10]
```

```python
fifa.club.value_counts()[:10].plot(kind='bar', color='brown', figsize=(14, 6))
```

```python
fifa.groupby('club')['wage_eur'].mean()
```

```python
average_club_wage_expenses = fifa.groupby('club')['wage_eur'].sum()
average_club_wage_expenses.sort_values(ascending=False)[:10]
```

```python
average_club_wage_expenses_t10 = average_club_wage_expenses.sort_values(ascending=False)[:10]
```

```python
average_club_wage_expenses_t10.index
```

```python
average_club_wage_expenses_t10.values
```

```python
fig = plt.figure(figsize=(20, 10))
sns.barplot(x=average_club_wage_expenses_t10.index, y=average_club_wage_expenses_t10.values)
plt.title('10 Clubs that spend highest on player wages', fontsize=20, weight = 'bold')
plt.xlabel('CLUB', fontsize=20, weight='bold')
plt.ylabel('AVERAGE WAGE', fontsize=20, weight='bold')
plt.xticks(weight='bold', fontsize=15, rotation=90)
plt.yticks(weight='bold', fontsize=15)
plt.grid()
plt.show()
```

```python
average_club_wage_expenses_low10 = average_club_wage_expenses.sort_values(ascending=True)[:10]
average_club_wage_expenses_low10
```
